```python
# Image Filter App for Google Colab
from google.colab import files
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
import io


# Upload image
print("üì§ Upload an image:")
uploaded = files.upload()


# Load the image
image_name = list(uploaded.keys())[0]
image = Image.open(image_name)


# Display original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis("off")


# Apply filters
def apply_filter(image, filter_type):
    if filter_type == "grayscale":
        return image.convert("L")
    elif filter_type == "sepia":
        img = np.array(image)
        tr = [0.393, 0.769, 0.189]
        tg = [0.349, 0.686, 0.168]
        tb = [0.272, 0.534, 0.131]


        sepia = np.dot(img[..., :3], [tr, tg, tb])
        sepia = np.clip(sepia, 0, 255)
        sepia = sepia.astype("uint8")


        return Image.fromarray(sepia)
    elif filter_type == "blur":
        return image.filter(ImageFilter.BLUR)
    elif filter_type == "edge":
        return image.filter(ImageFilter.FIND_EDGES)
    else:
        return image


# Choose filter
print("\nChoose a filter:")
print("1. Grayscale")
print("2. Sepia")
print("3. Blur")
print("4. Edge Detection")


choice = input("Enter filter number (1-4): ")


filters = {
    "1": "grayscale",
    "2": "sepia",
    "3": "blur",
    "4": "edge"
}


if choice in filters:
    filtered_image = apply_filter(image, filters[choice])


    # Display filtered image
    plt.subplot(1, 2, 2)
    plt.title(f"{filters[choice].capitalize()} Filter")
    plt.imshow(filtered_image)
    plt.axis("off")
    plt.show()


    # Save and download filtered image
    output_name = f"filtered_{filters[choice]}.png"
    filtered_image.save(output_name)
    files.download(output_name)


    print(f"‚úÖ Filter applied! Downloading '{output_name}'...")


else:
    print("‚ùå Invalid choice. Try again!")
```
